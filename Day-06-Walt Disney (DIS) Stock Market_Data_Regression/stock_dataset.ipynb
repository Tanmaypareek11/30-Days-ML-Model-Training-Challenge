```python
# 1. Create lag features
df['Close_lag1'] = df['Close'].shift(1)
df['Close_lag2'] = df['Close'].shift(2)

# 2. Create target (future prediction)
df['Target'] = df['Close'].shift(-1)

# 3. Remove NaN values
df = df.dropna()

# 4. Define features & target
x = df[['Close_lag1', 'Close_lag2', 'Volume']]
y = df['Target']
```

```python
x
```

```python
# train test split 
# we use this method because if we use the train_test_split than it shuffles data randomly and the future data go into the trining part and the past data into the testing part
x_train, x_test = x[:int(0.8*len(x))], x[int(0.8*len(x)):]
y_train, y_test = y[:int(0.8*len(y))], y[int(0.8*len(y)):]

```

```python
# Here we are shifting the data because "To predict today’s price, look at previous days’ prices"
df['Close_lag1'] = df['Close'].shift(1)
df['Close_lag2'] = df['Close'].shift(2)
```

### Linear Regression Model 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("MAE:", mae)
print("MSE:",mse)
print("R2 Score:", r2)
print("RMSE:", rmse)
```

### Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100,max_depth = 10, random_state=42)
rfr.fit(x_train,y_train)
```

```python
y_pred = rfr.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("MAE:", mae)
print("MSE:",mse)
print("R2 Score:", r2)
print("RMSE:", rmse)
```
