```python
# feature engineering 
df['team_avg_score'] = df.groupby('batting_team')['total_runs'].transform('mean')
df['venue_avg_score'] = df.groupby('venue')['total_runs'].transform('mean')
df['toss_win'] = (df['toss_winner'] == df['batting_team']).astype(int)
df['season_avg_score'] = df.groupby('season')['total_runs'].transform('mean')
df['team_venue_avg'] = df.groupby(['batting_team','venue'])['total_runs'].transform('mean')
df['city_avg_score'] = df.groupby('city')['total_runs'].transform('mean')
df['city_avg_score'].fillna(df['venue_avg_score'], inplace=True)
```

```python
#encoding 
from sklearn.preprocessing import LabelEncoder

X_cat = pd.get_dummies(df[['batting_team','venue','city','toss_decision']], drop_first=True)
X_num = df[['team_avg_score','venue_avg_score','season_avg_score',
            'team_venue_avg','city_avg_score','toss_win']]

X = pd.concat([X_cat, X_num], axis=1)

y = df['total_runs']
```

```python
#train_test_split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    X,y,test_size = 42,random_state = 42
)
```

```python
X.isnull().sum()
```

# model training
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train,y_train)

```python
y_pred = linreg.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
