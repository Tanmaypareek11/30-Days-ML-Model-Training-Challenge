```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_csv(r"C:\Users\tanma\Downloads\DATASETS\cars_dataset\car_price_dataset.csv")
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.drop("Car_ID",axis =1)
```

```python
df
```

```python
# Encoding of the categorical columns by using one hot encoding 
x = df.drop("Price",axis =1)
y =df["Price"]
x = pd.get_dummies(x,drop_first=True)
```

```python
x
```

```python
# split the data 
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state = 42
)
```

### Linear Regression Model 

```python
# linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_pred,y_test)
MAE = mean_absolute_error(y_pred,y_test)
MSE = mean_squared_error(y_pred,y_test)
rmse = mean_squared_error(y_pred,y_test,squared=False)
print("R2 score is :",r2)
print("mean_absolute_error is:",MAE)
print("mean_squared_error is :",MSE)
print("RMSE :",rmse)
```

### XGBoost Model 

```python
# XGBoost 
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=200,max_depth=5,random_state=42)
```

```python
xgb.fit(x_train,y_train)
```

```python
y_pred = xgb.predict(x_test)
```

```python
r2 = r2_score(y_pred,y_test)
MAE = mean_absolute_error(y_pred,y_test)
MSE = mean_squared_error(y_pred,y_test)
rmse = mean_squared_error(y_pred,y_test,squared=False)
print("R2 score is :",r2)
print("mean_absolute_error is:",MAE)
print("mean_squared_error is :",MSE)
print("RMSE :",rmse)
```

```python
# hyperparameter tuning in the XGBoost 
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 300, 500],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 0.9, 1]
}
```

```python
XGB = XGBRegressor(random_state=42)

grid = GridSearchCV(
    estimator=XGB,
    param_grid=param_grid,
    cv=5
)

grid.fit(x_train, y_train)
```

```python
print("Best Parameters:", grid.best_params_)
```

```python
best_model = grid.best_estimator_

y_pred = best_model.predict(x_test)
```

```python
r2 = r2_score(y_pred,y_test)
MAE = mean_absolute_error(y_pred,y_test)
MSE = mean_squared_error(y_pred,y_test)
rmse = mean_squared_error(y_pred,y_test,squared=False)
print("R2 score is :",r2)
print("mean_absolute_error is:",MAE)
print("mean_squared_error is :",MSE)
print("RMSE :",rmse)
```
